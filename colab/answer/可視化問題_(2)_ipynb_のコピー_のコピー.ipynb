## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
#データ読み込み
data = pd.read_csv('hist_data.csv')
#aデータ　bデータに分ける
x = data['aデータ']
y = data['bデータ']

#それぞれのヒストグラムの設定　bins は度数(今回は手本グラフから30)
#alphaはメイド
plt.hist(x, bins=30, label='aデータ', color='yellow', alpha=0.5)
plt.hist(y, bins=30, label='bデータ', color='green', alpha=0.5)

plt.xlabel('温度(℃)',)
plt.ylabel('度数')
plt.title('aデータとbデータの温度に関する度数分布')
#わかりやすくするため例を追加
plt.legend()
plt.show()

```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
#データの読み込み
data = pd.read_csv('scatter_data.csv')
#データの分割
data_length = data["length (cm)"]
data_width = data["width (cm)"]
data_class = data["class"]

plt.title('クラス０～２のlengthとwidthに関する散布図')
plt.xlabel("length(cm)")
plt.ylabel("width(cm)")
#散布図の作成
plt.scatter(data_length[data_class==0], data_width[data_class==0], color='red', label='class0')
plt.scatter(data_length[data_class==1], data_width[data_class==1], color='blue', label='class1')
plt.scatter(data_length[data_class==2], data_width[data_class==2], color='green', label='class2')

plt.grid(True)
plt.legend()
plt.show()
```

```python
#(1)
data = pd.read_csv('line_data.csv')
x = data['期間']
y = data['費用']
z = data['件数']

plt.title('期間ごとの件数の推移')
plt.xlabel('期間')
plt.ylabel('件数)')
#折れ線グラフの設定
plt.plot(x, z, marker='o', linestyle='-', color='blue')
#マス目の設定
plt.grid(True)
#横軸の文字の大きさ調整
plt.xticks(rotation=45, fontsize=8)
plt.show()

#(2)
plt.title('期間ごとの費用の推移')
plt.xlabel('期間')
plt.ylabel('費用（円)')
#棒グラフの作成
plt.bar(x, y)
plt.xticks(rotation=45, fontsize=8)
plt.grid(True)


plt.show()

```
